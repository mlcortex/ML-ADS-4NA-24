## Vamos gerar algumas estatísticas básicas sobre a base:


```python
#média
#antes eu preciso acessar uma coluna específica
#vamos calcular o comprimento médio da sépala
dados['sepal length'].mean()
```

```python
dados['class']
```

```python
#vamos fazer para todas as colunas
for nome_coluna in dados.columns:
  #nesse caso vamos pular se a coluna for clase por essa não ser uma numérica
  print(nome_coluna)
  if nome_coluna == 'class':
    print('pulei essa coluna')
    continue
  valor_medio = dados[nome_coluna].mean()
  print(f'O valor médio da coluna {nome_coluna} é {valor_medio}')
```

```python
# eu ainda preciso fazer o mesmo para outras estatísticas descritivas:
# quais são elas? Mediana, máximo, mínimo, desvio padrão (std)
```

```python
# prompt: #vamos fazer para todas as colunas
# for nome_coluna in dados.columns:
#   #nesse caso vamos pular se a coluna for clase por essa não ser uma numérica
#   print(nome_coluna)
#   if nome_coluna == 'class':
#     print('pulei essa coluna')
#     continue
#   valor_medio = dados[nome_coluna].mean()
#   print(f'O valor médio da coluna {nome_coluna} é {valor_medio}')
# expandir o cod anterior para media mediana e afins

#vamos fazer para todas as colunas
for nome_coluna in dados.columns:
  #nesse caso vamos pular se a coluna for clase por essa não ser uma numérica
  print(nome_coluna)
  if nome_coluna == 'class':
    print('pulei essa coluna')
    continue
  valor_medio = dados[nome_coluna].mean()
  valor_mediana = dados[nome_coluna].median()
  valor_maximo = dados[nome_coluna].max()
  valor_minimo = dados[nome_coluna].min()
  valor_desvio_padrao = dados[nome_coluna].std()
  print(f'O valor médio da coluna {nome_coluna} é {valor_medio}')
  print(f'O valor da mediana da coluna {nome_coluna} é {valor_mediana}')
  print(f'O valor máximo da coluna {nome_coluna} é {valor_maximo}')
  print(f'O valor mínimo da coluna {nome_coluna} é {valor_minimo}')
  print(f'O valor do desvio padrão da coluna {nome_coluna} é {valor_desvio_padrao}')
```

```python
# e como fazemos para valores categóricos, tipo a class?
# aqui é contagem dos valores mesmo
dados['class'].value_counts()
```

```python
#isolando apenas as classes únicas:
print(dados['class'].unique())
print(f'temos {len(dados["class"].unique())} classes')
```

Com isso podemos entender que temos 3 classes Iris-setosa Iris-versicolor e Iris-virginica.

Cada classe com 50 exemplos observados.
