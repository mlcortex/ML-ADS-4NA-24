2. Criação de Boxplots Crie boxplots para explorar a dispersão dos dados nas seguintes colunas:   
Dica: Use bibliotecas como `matplotlib` ou seaborn para criar gráficos de boxplot. As funções `sns.boxplot()` e `plt.boxplot()` podem ser úteis.
* magnesio
* prolina
* cinzas
* alcool

````python
import seaborn as sns
import matplotlib.pyplot as plt

for coluna in ['magnesio','prolina','cinzas','alcool']:
  plt.figure(figsize=(15, 6))
  sns.boxplot(vinhos[coluna],orient='h',width=.5,flierprops={"marker": "x"})
  plt.show()
````
