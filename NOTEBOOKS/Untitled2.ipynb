```python
# Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

# 1. Análise Exploratória de Variáveis

# Contagem das diferentes classes presentes no dataset
classe_counts = vinhos['classe'].value_counts()
print("Contagem de Classes:\n", classe_counts)

# Estatísticas descritivas para as colunas especificadas
estatisticas = vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe()
print("\nEstatísticas Descritivas:\n", estatisticas)

# 2. Criação de Boxplots

# Configurar o estilo dos gráficos
sns.set(style="whitegrid")

# Criar boxplots para explorar a dispersão dos dados
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.boxplot(y=vinhos['magnesio'])
plt.title('Boxplot de Magnésio')
plt.ylabel('Magnésio')

plt.subplot(2, 2, 2)
sns.boxplot(y=vinhos['prolina'])
plt.title('Boxplot de Prolina')
plt.ylabel('Prolina')

plt.subplot(2, 2, 3)
sns.boxplot(y=vinhos['cinzas'])
plt.title('Boxplot de Cinzas')
plt.ylabel('Cinzas')

plt.subplot(2, 2, 4)
sns.boxplot(y=vinhos['alcool'])
plt.title('Boxplot de Álcool')
plt.ylabel('Álcool')

plt.tight_layout()
plt.show()

# 3. Boxplots Segregados por Classe

# Criar boxplots segregados pela variável classe
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
vinhos.boxplot(column='magnesio', by='classe')
plt.title('Boxplot de Magnésio por Classe')
plt.suptitle('')
plt.ylabel('Magnésio')

plt.subplot(2, 2, 2)
vinhos.boxplot(column='fenois_totais', by='classe')
plt.title('Boxplot de Fenois Totais por Classe')
plt.suptitle('')
plt.ylabel('Fenois Totais')

plt.subplot(2, 2, 3)
vinhos.boxplot(column='acido_malico', by='classe')
plt.title('Boxplot de Ácido Málico por Classe')
plt.suptitle('')
plt.ylabel('Ácido Málico')

plt.subplot(2, 2, 4)
vinhos.boxplot(column='alcool', by='classe')
plt.title('Boxplot de Álcool por Classe')
plt.suptitle('')
plt.ylabel('Álcool')

plt.tight_layout()
plt.show()

# 4. Criação de Histogramas

# Histograma da coluna fenois_totais
plt.figure(figsize=(10, 5))
plt.hist(vinhos['fenois_totais'], bins=10, color='blue', alpha=0.7)
plt.title('Histograma de Fenois Totais')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.grid()
plt.show()

# Histograma da coluna fenois_totais filtrando apenas a classe 3
plt.figure(figsize=(10, 5))
plt.hist(vinhos[vinhos['classe'] == '3']['fenois_totais'], bins=10, color='red', alpha=0.7)
plt.title('Histograma de Fenois Totais - Classe 3')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.grid()
plt.show()
```
