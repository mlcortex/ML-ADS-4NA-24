 # Importando bibliotecas necessarias para obter os dados requisitados

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Importando a planilha traduzida em português

```python

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Lendo os dados da coluna
df = pd.read_csv(url)

# Atribuindo os dados as novas colunas
df.columns = column_names_pt_br

#Convertendo 'classe' para o tipo string
df['classe'] = df['classe'].astype(str)
```

#### Contagem das classes conditas:


```python
contagem = df['classe'].value_counts()
print("Contagem das classes:")
print(contagem)
```

#### Estatísticas descritivas


```python
estatistica = df[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe()
print("Estatísticas descritivas:")
print(estatistica)
```

#### Insights


```python
print("Insights:")
for ins in ['alcool', 'acido_malico', 'magnesio', 'matiz']:
    media = df[ins].mean()
    mediana = df[ins].median()
    std_dev = df[ins].std()
    min_val = df[ins].min()
    max_val = df[ins].max()

    print(f"- {ins}:")
    print(f"  Média: {media:.2f}, Mediana: {mediana:.2f}, Desvio Padrão: {std_dev:.2f}")
    print(f"  Mínimo: {min_val:.2f}, Máximo: {max_val:.2f}")
```

# 2. Criação de Boxplots
Crie boxplots para explorar a dispersão dos dados nas seguintes colunas:
magnesio, prolina, cinzas e alcool.

```python
#Magnésio
plt.subplot(2, 1, 2)
sns.boxplot(x=df['magnesio'])
plt.title('Boxplot de Magnésio')
```

```python
#Prolina
plt.subplot(2, 1, 1)
sns.boxplot(x=df['prolina'])
plt.title('Boxplot de Prolina')
```

```python
#Cinzas
plt.subplot(2, 1, 2)
sns.boxplot(x=df['cinzas'])
plt.title('Boxplot de Cinzas')
```

```python
#Álcool
plt.subplot(2, 1, 2)
sns.boxplot(x=df['alcool'])
plt.title('Boxplot de Álcool')
```

# 3. Boxplots Segregados por Classe
Crie um boxplot das seguintes colunas, segregando os dados pela variável classe:
magnesio, fenois_totais, acido_malico e alcool.

```python
colunas = ['magnesio', 'fenois_totais', 'acido_malico', 'alcool']
for coluna in colunas:
    plt.figure(figsize=(10, 6))
    df.boxplot(column=coluna, by='classe')
    plt.title(f'Boxplot de {coluna.capitalize()} por Classe')
    plt.suptitle('')
    plt.ylabel(coluna)
    plt.show()
```

# 4. Criação de Histograma


Crie um histograma para visualizar a distribuição dos dados na coluna

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['fenois_totais'], bins=10, kde=True)
plt.title('Histograma de Fenóis Totais')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

Crie um histograma da coluna fenois_totais, filtrando apenas a classe 3, utilizando 10 bins.

```python
vinhos_3 = df[df['classe'] == '3']

plt.figure(figsize=(10, 6))
sns.histplot(vinhos_3['fenois_totais'], bins=10, kde=True)
plt.title('Histograma de Fenóis Totais (Classe 3)')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```
