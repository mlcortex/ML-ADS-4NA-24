Comecei fazendo a leitura dos dados usando o código fornecido pelo professor:

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

#Questão 02
- Crie boxplots para explorar a dispersão dos dados nas seguintes colunas:

- Magnesio

```python
plt.figure(figsize=(4, 4))
sns.boxplot(y=vinhos['magnesio'])
plt.title('Boxplot da coluna Magnesio')
plt.ylabel('Magnesio')
plt.show()
```

- Prolina

```python
plt.figure(figsize=(4, 4))
sns.boxplot(y=vinhos['prolina'])
plt.title('Boxplot da coluna Prolina')
plt.ylabel('Prolina')
plt.show()
```

- Cinzas

```python
plt.figure(figsize=(4, 4))
sns.boxplot(y=vinhos['cinzas'])
plt.title('Boxplot da coluna Cinzas')
plt.ylabel('Cinzas')
plt.show()
```

- Alcool

```python
plt.figure(figsize=(4, 4))
sns.boxplot(y=vinhos['alcool'])
plt.title('Boxplot da coluna Alcool')
plt.ylabel('Alcool')
plt.show()
```

#Questão 04
Crie um histograma para visualizar a distribuição dos dados na coluna:

```python
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']

plt.hist(vinhos['fenois_totais'], bins=10, edgecolor='black')

plt.title('Histograma de Fenois Totais')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')

plt.show()
```
