```python
import pandas as pd;
import matplotlib.pyplot as plt;
import seaborn as sns;
```

Importando bibliotecas

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data';

column_names_pt_br = [
    'classe', 'alcool', 'acido_malico', 'cinzas',
    'alcalinidade_de_cinzas', 'magnesio', 'fenois_totais',
    'flavanoides', 'fenois_nao_flavanoides', 'proantocianinas',
    'intensidade_de_cor', 'matiz', 'od280_od315_de_vinhos_diluidos',
    'prolina'
];
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object});
print(vinhos);
```

Lendo os dados.

```python
contagem_classes = vinhos['classe'].value_counts();
print(contagem_classes);

estatisticas = vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe();
print(estatisticas);
```

Análise das classes.

Agora criaremos o bloxpot

```python
sns.set(style="whitegrid");

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.boxplot(y=vinhos['magnesio'])
plt.title('Boxplot de Magnésio')
plt.ylabel('Magnésio')

plt.subplot(2, 2, 2)
sns.boxplot(y=vinhos['prolina'])
plt.title('Boxplot de Prolina')
plt.ylabel('Prolina')

plt.subplot(2, 2, 3)
sns.boxplot(y=vinhos['cinzas'])
plt.title('Boxplot de Cinzas')
plt.ylabel('Cinzas')

plt.subplot(2, 2, 4)
sns.boxplot(y=vinhos['alcool'])
plt.title('Boxplot de Álcool')
plt.ylabel('Álcool')

plt.tight_layout()
plt.show()
```

Boxplots gerados

Boxplots segregados por classe:

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

vinhos.boxplot(column='magnesio', by='classe', ax=axs[0, 0])
axs[0, 0].set_title('Boxplot de Magnésio por Classe')
axs[0, 0].set_ylabel('Magnésio')
axs[0, 0].set_xlabel('Classe')

vinhos.boxplot(column='fenois_totais', by='classe', ax=axs[0, 1])
axs[0, 1].set_title('Boxplot de Fenóis Totais por Classe')
axs[0, 1].set_ylabel('Fenóis Totais')
axs[0, 1].set_xlabel('Classe')

vinhos.boxplot(column='acido_malico', by='classe', ax=axs[1, 0])
axs[1, 0].set_title('Boxplot de Ácido Málico por Classe')
axs[1, 0].set_ylabel('Ácido Málico')
axs[1, 0].set_xlabel('Classe')

vinhos.boxplot(column='alcool', by='classe', ax=axs[1, 1])
axs[1, 1].set_title('Boxplot de Álcool por Classe')
axs[1, 1].set_ylabel('Álcool')
axs[1, 1].set_xlabel('Classe')

plt.tight_layout()
plt.suptitle('')
plt.show()
```

Agora iremos para os histogramas

```python
plt.figure(figsize=(10, 5))
sns.histplot(vinhos['fenois_totais'], bins=10, kde=True)
plt.title('Histograma de Fenóis Totais')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(vinhos[vinhos['classe'] == '3']['fenois_totais'], bins=10, kde=True)
plt.title('Histograma de Fenóis Totais - Classe 3')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.show()
```
