```
# !pip install seaborn matplotlib

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style='whitegrid')


url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'


column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]


vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

# 1. Análise Exploratória de Variáveis

## Contagem das classes presentes no dataset
print("Contagem das classes:")
print(vinhos['classe'].value_counts())

## Estatísticas descritivas das colunas solicitadas
print("\nEstatísticas descritivas de 'alcool':")
print(vinhos['alcool'].describe())

print("\nEstatísticas descritivas de 'acido_malico':")
print(vinhos['acido_malico'].describe())

print("\nEstatísticas descritivas de 'magnesio':")
print(vinhos['magnesio'].describe())

print("\nEstatísticas descritivas de 'matiz':")
print(vinhos['matiz'].describe())

# 2. Criação de Boxplots

## Boxplot para a coluna 'magnesio'
plt.figure(figsize=(8, 6))
sns.boxplot(data=vinhos, y='magnesio')
plt.title('Boxplot de Magnésio')
plt.show()

## Boxplot para a coluna 'prolina'
plt.figure(figsize=(8, 6))
sns.boxplot(data=vinhos, y='prolina')
plt.title('Boxplot de Prolina')
plt.show()

## Boxplot para a coluna 'cinzas'
plt.figure(figsize=(8, 6))
sns.boxplot(data=vinhos, y='cinzas')
plt.title('Boxplot de Cinzas')
plt.show()

## Boxplot para a coluna 'alcool'
plt.figure(figsize=(8, 6))
sns.boxplot(data=vinhos, y='alcool')
plt.title('Boxplot de Álcool')
plt.show()

# 3. Boxplots Segregados por Classe

## Boxplot de 'magnesio' por classe
vinhos.boxplot(column='magnesio', by='classe', figsize=(8, 6))
plt.title('Boxplot de Magnésio por Classe')
plt.suptitle('')  # Remove o título automático do matplotlib
plt.show()

## Boxplot de 'fenois_totais' por classe
vinhos.boxplot(column='fenois_totais', by='classe', figsize=(8, 6))
plt.title('Boxplot de Fenois Totais por Classe')
plt.suptitle('')
plt.show()

## Boxplot de 'acido_malico' por classe
vinhos.boxplot(column='acido_malico', by='classe', figsize=(8, 6))
plt.title('Boxplot de Ácido Málico por Classe')
plt.suptitle('')
plt.show()

## Boxplot de 'alcool' por classe
vinhos.boxplot(column='alcool', by='classe', figsize=(8, 6))
plt.title('Boxplot de Álcool por Classe')
plt.suptitle('')
plt.show()

# 4. Criação de Histograma

## Histograma para a coluna 'fenois_totais' com 10 bins
plt.figure(figsize=(8, 6))
plt.hist(vinhos['fenois_totais'], bins=10, edgecolor='black')
plt.title('Histograma de Fenois Totais')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()

## Histograma para 'fenois_totais' filtrando apenas a classe 3
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']
plt.figure(figsize=(8, 6))
plt.hist(vinhos_classe_3['fenois_totais'], bins=10, edgecolor='black')
plt.title('Histograma de Fenois Totais (Classe 3)')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```
