```python

import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

#Terceira questão

Classe magnesio

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14, 10))
```

```python
plt.subplot(2, 2, 1)
vinhos.boxplot(column='magnesio', by='classe', grid=False)
plt.title('Boxplot - Magnésio por Classe')
plt.suptitle('')
plt.xlabel('Classe')
plt.ylabel('Magnésio')
```

```python
plt.subplot(2, 2, 2)
vinhos.boxplot(column='fenois_totais', by='classe', grid=False)
plt.title('Boxplot - Fenois Totais por Classe')
plt.suptitle('')
plt.xlabel('Classe')
plt.ylabel('Fenois Totais')
```

#Quarta questão

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(vinhos['fenois_totais'], bins=10, kde=False)
plt.title('Histograma de Fenois Totais')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')


plt.show()
```

```python
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']


plt.figure(figsize=(8, 6))
sns.histplot(vinhos_classe_3['fenois_totais'], bins=10, kde=False)

plt.title('Histograma de Fenois Totais - Classe 3')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')


plt.show()
```
